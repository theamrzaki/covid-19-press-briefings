```
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```
def parse_breefing(main_counter,page_link = "https://www.rev.com/blog/transcripts/governor-gavin-newsom-california-covid-19-briefing-transcript-may-8"):
    page = requests.get(page_link)
    soup = BeautifulSoup(page.content, 'html.parser')
    press = soup.find_all('div', class_='fl-callout-text')
    date = soup.find_all("div",class_="fl-rich-text")
    header = soup.find_all("h1",class_="fl-heading")
    for i,a in enumerate(press[0].children):
      try:
        link = a.find("a")
        time = link.text
        text = a.text.replace("("+link.text+")","")
        speaker = text.split(":")[0]
        text = "".join(text.split(":")[1])
        output.append({"speaker":speaker ,
                        "text":text.strip(),
                      "time": time,
                      "transcript_link":link["href"],
                      "date":date[0].text,
                      "header":header[0].text.strip(),
                      "page_link":page_link,
                       "main_page_counter":main_counter})
      except Exception as e:
        print(str(e)+ " " + page_link + " in " + str(a))
```

```
output = []
main_counter = 0 
main_link = "https://www.rev.com/blog/transcript-tag/coronavirus-update-transcripts/page/"
for p in tqdm(list(range(1,60+1))):
  main_page = requests.get("https://www.rev.com/blog/transcript-tag/coronavirus-update-transcripts/page/"+str(p))
  soup_0 = BeautifulSoup(main_page.content, 'html.parser')
  main_links = [p.find("a")["href"] for p in soup_0.find_all('div', class_='fl-post-column')]
  for m in main_links:
    parse_breefing(main_counter,m)
    main_counter+=1
```

```
from collections import Counter 
from tqdm import tqdm 

def get_speaker_identity(speaker,text_per_speaker):
  speaker_identity = {}
  x = Counter(speaker) 
  all_speakers = list(x.keys())
  operator = speaker[0]
  main_speaker = x.most_common(1)[0][0]
  speaker_identity[operator] = "operator"
  speaker_identity[main_speaker] = "main_speaker"

  for t in text_per_speaker[main_speaker]:
    for speak in all_speakers:
      #speak = speak.split(".")[-1]
      if str(speak).lower() in str(t).lower():
        #print(speak)
        speaker_identity[speak] = "helper"

  for speak in all_speakers:
    if speak not in speaker_identity.keys():
      speaker_identity[speak]="reporter"
  return speaker_identity
```

```
output_df_grouped_with_identity = []
questions = []
answers = []
time = []
transcript_link_list = []
date=[]
header=[]
page_link=[]
main_page_counter=[]
speaker = []
speaker_type = []
output_df_grouped = output_df.groupby("main_page_counter")
text_per_speaker = {}
limit_question_words = 10
for i, group in tqdm(output_df_grouped):
  link = list(group["page_link"])
  speaker = list(group["speaker"])
  text = list(group["text"])

  #prepare to calculate identity
  for i,speak in enumerate(speaker):
    if speak not in text_per_speaker.keys():
      text_per_speaker[speak] = []
    text_per_speaker[speak].append(text[i])
  speaker_identity = get_speaker_identity(speaker,text_per_speaker)

  #append identity
  identity = []
  for i,speak in enumerate(speaker):
    identity.append(speaker_identity[speak])
  group["identity"] = identity
  output_df_grouped_with_identity.append(group)

  #get Q/A
  start_index = 0
  index=group.index[0]
  while True:
    if index >= group.index[-1]:
      break
    current_speaker = group.loc[index]["speaker"]
    current_question = ""
    if group.loc[index]["identity"] == "reporter":
      start_index = index
      current_question += str(group.loc[index]["text"])
      while True:
        if index+1 <= group.index[-1] and group.loc[index+1]["speaker"] == current_speaker :
          current_question += str(group.loc[index+1]["text"])
          if len(current_question.split(" "))<limit_question_words:#break on short questions
            break
          index+=1
        else:
          if len(current_question.split(" "))<limit_question_words:#break on short questions
            break
          answer = str(group.loc[index+1]["text"]) if index+1 <= group.index[-1] else ""
          if index+2 <= group.index[-1] and group.loc[index+2]["identity"] in ["main_speaker","helper","operator"]:
            answer+= str(group.loc[index+2]["text"])
            index+=1
          if len(answer.split(" "))<limit_question_words:#break on short answers
            break
          questions.append(current_question)
          answers.append(answer)
          time.append(group.loc[start_index]["time"])
          transcript_link_list.append(group.loc[start_index]["transcript_link"])
          date.append(group.loc[start_index]["date"])
          header.append(group.loc[start_index]["header"])
          main_page_counter.append(group.loc[start_index]["main_page_counter"])
          break
    index+=1
  #break
```

```
df_identity = pd.concat(output_df_grouped_with_identity)
df_identity.to_csv("/content/drive/My Drive/COVID Press/covid_press_identity.csv",index=False)
df_identity.head()
```

```
QA_press = pd.DataFrame({ "questions":questions,
                          "answers":answers,
                          "time":time,
                          "transcript_link":transcript_link_list,
                          "date":date,
                          "header":header,
                          "main_page_counter":main_page_counter})
QA_press.to_csv("/content/drive/My Drive/COVID Press/covid_press_QA.csv",index=False)
QA_press.head()
```
